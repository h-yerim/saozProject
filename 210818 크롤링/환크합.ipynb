```python
import pandas as pd
a= pd.read_csv('환경부 크롤링1629249364')
b= pd.read_csv('환경부 크롤링1629252901')
c= pd.read_csv('환경부 크롤링1629256500')
d= pd.read_csv('환경부 크롤링1629260104')
e= pd.read_csv('환경부 크롤링1629265986')
f= pd.read_csv('환경부 크롤링1629275359')
g= pd.read_csv('환경부 크롤링1629280529')
h= pd.read_csv('환경부 크롤링1629288030')
i= pd.read_csv('환경부 크롤링1629291963')
```

```python
x=pd.read_excel('data/유선부분_동.xlsx')
x=x[['동']]
y=pd.read_excel('data/승현.xlsx')
y=y[['동']]
xy=pd.concat([y,x],ignore_index=True)
xy
```

```python
df_list=[a,b,c,d,e,f,g,h,i]
dfs=[]
for x in df_list:
    x=x[['구명','시설분류','주차장명','사용상태','충전기타입','수집시간']]
    x['동']=xy['동']
    dfs.append(x)
```

```python
result=pd.DataFrame(columns=(['구명','시설분류','주차장명','사용상태','충전기타입','수집시간']))
result=pd.concat([dfs[0],dfs[1],dfs[2],dfs[3],dfs[4],dfs[5],dfs[6],dfs[7],dfs[8]],ignore_index=True)
result.head()
```

```python
result=result.sort_values(by=['구명','주차장명','수집시간'],ascending=[True,True,True])
result=result.reset_index(drop=True)
result.head()
```

```python
result.to_csv('210818 환경부 크롤링.csv',index=False)
```
